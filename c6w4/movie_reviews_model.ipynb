```python
import nltk
```

```python
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
```

```python
import pickle
```

```python
negids = movie_reviews.fileids('neg')
posids = movie_reviews.fileids('pos')
```

```python
negfeats = [movie_reviews.words(fileids=[x]) for x in negids]
posfeats = [movie_reviews.words(fileids=[x]) for x in posids]
```

```python
reviews = []

for i in negfeats:
  reviews.append(' '.join(i))

for j in posfeats:
  reviews.append(' '.join(j))
```

```python
import numpy as np
```

```python
y = np.array([0] *  1000 + [1] * 1000)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import cross_val_score
```

```python
pipe = Pipeline([ ('vectorizer', CountVectorizer()), ('classifier', LogisticRegression()) ])
```

```python
cross_val_score(pipe, reviews, y, cv=5).mean()
```

Сохраняем обученную модель векторайзера:

```python
vectorizer = CountVectorizer().fit(reviews)
```

```python
X = vectorizer.transform(reviews)
```

Сохраняем обученную модель классификатора:

```python
classifier = LogisticRegression().fit(X, y)
```

```python
def proba(vectorized):
  return classifier.predict_proba(vectorized)[0].max()
```

Проверим работу векторизатора и классификатора на пробных текстах. Оба текста - обзоры на фильм "Однажды в Америке". Сначала проверим на положительном отзыве с оценкой 10:

```python
good_text = 'It goes without saying that the cast is wonderful, and packed with the usual suspects doing their thing. De Niro excels and has the best old-age makeup I\'ve ever seen, James Woods is on the edge as usual, and Elizabeth McGovern is Deborah, an intense object of desire for De Niro. Did anyone ever think that the reason she hasn\'t aged at all at the end of the film is because we\'re seeing reality through De Niro\'s eyes, and in his eyes she is still as young and beautiful as she ever was? Anyway, plenty of familiar faces pack out the rest of the cast, from Treat Williams, Joe Pesci, and Burt Young to Danny Aiello as the hilarious police chief. William Forsythe, Darlanne Fluegel, the list is endless and also includes Jennifer Connelly in her first acting performance.'.lower()
```

```python
vectorized_good = vectorizer.transform([good_text])
```

```python
score_good = classifier.predict(vectorized_good)
```

```python
score_good[0]
```

```python
proba(vectorized_good)
```
