```python
from nltk.corpus import movie_reviews
```

```python
import nltk
```

```python
nltk.download('movie_reviews')
```

```python
negids = movie_reviews.fileids('neg')
```

```python
posids = movie_reviews.fileids('pos')
```

```python
len(negids), len(posids)
```

```python
negfeats = [movie_reviews.words(fileids=[x]) for x in negids]
```

```python
posfeats = [movie_reviews.words(fileids=[x]) for x in posids]
```

```python
f'Общее количество отзывов: {len(posfeats) + len(negfeats)}'
```

```python
f'Доля позитивных отзывов: {len(posfeats) / (len(posfeats) + len(negfeats))}'
```

```python
import numpy as np
```

```python
y = np.array([0] * 1000 + [1] * 1000)
```

```python
y.shape
```

```python
allfeats = []
```

```python
for i in negfeats:
  allfeats.append(' '.join(i))

for j in posfeats:
  allfeats.append(' '.join(j))
```

```python
len(allfeats)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(allfeats)
```

```python
  f'Количество признаков: {X.shape[1]}'
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_list = cross_val_score(logreg, X, y, scoring='accuracy')
```

```python
accuracy_list
```

```python
f'Accuracy в кросс-валидации {round(np.mean(accuracy_list), 2)}'
```

```python
roc_auc_list = cross_val_score(logreg, X, y, scoring='roc_auc')
```

```python
roc_auc_list
```

```python
f'ROC-AUC в кросс-валидации: {round(np.mean(roc_auc_list), 2)}'
```

```python
logreg2 = LogisticRegression(solver='liblinear')
```

```python
logreg2.fit(X, y)
```

```python
names = vectorizer.get_feature_names()
```

```python
len(names)
```

```python
values = sorted(logreg2.coef_[0])[:5] + sorted(logreg2.coef_[0])[-5:]
```

```python
values
```

```python
list_coef = list(logreg2.coef_[0])
```

```python
for x in values:
  ind = list_coef.index(x)
  print(names[ind], x)
```
